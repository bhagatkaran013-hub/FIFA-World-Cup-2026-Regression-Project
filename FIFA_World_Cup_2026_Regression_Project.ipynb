# K-Nearest Neighbor(KNN) Algorithm

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
x = df.drop("performance_score", axis=1)

y = df["performance_score"]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 0.25, random_state=0)
```

```python
#feature Scaling

from sklearn.preprocessing import StandardScaler
st_x= StandardScaler()
x_train= st_x.fit_transform(x_train)
x_test= st_x.transform(x_test)
```

```python
#Fitting K-NN regressor to the training set
regressor= KNeighborsRegressor(n_neighbors=5, metric='minkowski', p=2 )
# p=1: Uses Manhattan distance instead of Euclidean distance.
#p>2: Minkowski distance becomes more sensitive to large differences in feature values.
regressor.fit(x_train, y_train)
```

```python
#Predicting the test set result

y_pred=regressor.predict(x_test)
```

```python
#Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

print("MAE:", mae)
```

```python
# Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

print("MSE:", mse)
```

```python
#Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

print("RMSE:", rmse)
```

```python
#R² Score
r2 = r2_score(y_test, y_pred)

print("R² Score:", r2)
```

# Actual vs Predicted Plot

```python
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Performance Score")
plt.ylabel("Predicted Performance Score")
plt.title("Actual vs Predicted Performance Score")

plt.show()
```

The calculated metrics provide a comprehensive evaluation of the Decision Tree Regressor's performance on the dataset. A high R² score suggests that the model can explain a significant portion of the variance in the target variable, while low MAE, MSE, and RMSE values indicate accurate predictions.
